```python
import pandas as pd
import numpy as np
```

```python
vgdf = pd.read_csv('VideoGameSales2016_Cleaned.csv')
display(vgdf)
```

```python
vgnames = vgdf[['Name', 'Genre', 'Platform', 'Rating', 'Year_of_Release']].copy()
display(vgnames)
```

```python
vgnames.reset_index(level = 0, inplace = True)
display(vgnames)
```

```python
vgnames.to_csv("VideoGameNames.csv", columns = ['Name', 'Genre', 'Platform', 'Rating', 'Year_of_Release'], index_label = 'VideoGameKey')
```

```python
vgplatform = vgdf['Platform'].copy()
vgplatform.drop_duplicates(inplace = True)
vgplatform.reset_index(inplace = True, drop = True)
display(vgplatform)
```

```python
vgplatform.to_csv('Platform.csv', index_label = 'PlatformKey')
```

```python
vgdeveloper = vgdf['Developer'].copy()
display(vgdeveloper)
```

```python
vgdeveloper.drop_duplicates(inplace = True)
display(vgdeveloper)
```

```python
vgdeveloper.sort_values(na_position = 'first', inplace = True)
vgdeveloper.reset_index(drop = True, inplace = True)
display(vgdeveloper)
```

```python
vgdeveloper.to_csv('VideoGameDeveloper.csv', index_label = 'DeveloperKey')
```

```python
vgpublisher = vgdf['Publisher'].copy()
display(vgpublisher)
```

```python
vgpublisher.drop_duplicates(inplace = True)
vgpublisher.sort_values(na_position = 'first', inplace = True)
vgpublisher.reset_index(drop = True, inplace = True)
display(vgpublisher)
```

```python
vgpublisher.to_csv('VideoGamePublisher.csv', index_label = 'PublisherKey')
```

```python
vgdf['JoinedNames'] = vgdf['Name'] + vgdf['Platform'] + vgdf['Year_of_Release'].astype('str')
vgnames['JoinedNames'] = vgnames['Name'] + vgnames['Platform'] + vgnames['Year_of_Release'].astype('str')
display(vgnames)
```

```python
#display(vgnames.drop_duplicates(subset = ['JoinedNames']))
print(vgnames[vgnames.duplicated(['JoinedNames'])])
```

```python
vgKeys = vgdf.merge(vgnames, on = 'JoinedNames', how = 'outer')
display(vgKeys)
```

```python
vgKeys.drop(columns  = ['Name_x', 'Platform_x', 'Genre_x', 'Rating_x', 'Year_of_Release_x', 'Name_y', 'Platform_y', 'Genre_y', 'Rating_y', 'JoinedNames', 'Year_of_Release_y'], inplace = True)
display(vgKeys)
```

```python
vgKeys.rename(columns = {"index" : "NameKey"}, inplace = True)
display(vgKeys)
```

```python
vgpublisher = vgpublisher.to_frame()
vgpublisher.reset_index(level = 0, inplace = True)
display(vgpublisher)
```

```python
vgKeys = vgKeys.merge(vgpublisher, left_on = 'Publisher', right_on = 'Publisher', how = 'outer')
display(vgKeys)
```

```python
vgKeys.drop(columns = ['Publisher'], inplace = True)
vgKeys.rename(columns = {"index" : "PublisherKey"}, inplace = True)
display(vgKeys)
```

```python
display(vgKeys)
```

```python
vgdeveloper = vgdeveloper.to_frame()
vgdeveloper.reset_index(level = 0, inplace = True)
display(vgdeveloper)
```

```python
vgKeys = vgKeys.merge(vgdeveloper, on = 'Developer', how = 'outer')
display(vgKeys)
```

```python
vgKeys.drop(columns = ['Developer'], inplace = True)
vgKeys.rename(columns = {"index" : "DeveloperKey"}, inplace = True)
display(vgKeys)
```

```python
vgScore = vgKeys[['NameKey', 'PublisherKey', 'DeveloperKey', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count']].copy()
display(vgScore)
```

```python
vgScore.dropna(subset = ['Critic_Score', 'Critic_Count', 'User_Score', 'User_Count'], inplace = True)
display(vgScore)
```

```python
vgScore.to_csv('FactScore.csv', index = False)
```

```python
display(vgKeys)
```

```python
vgKeys.rename(columns = {"NA_Sales" : "NA", "EU_Sales" : "EU", "JP_Sales": "JP", "Other_Sales": "OT"}, inplace = True)
display(vgKeys)
```

```python
vgSales = vgKeys[['NameKey', 'PublisherKey', 'DeveloperKey', 'NA', 'EU', 'JP', 'OT']]
display(vgSales)
```

```python
vgSalesMelt = pd.melt(vgSales, id_vars = ['NameKey', 'PublisherKey', 'DeveloperKey'], value_vars = ['NA', 'EU', 'JP', 'OT'], value_name = 'Sales', var_name = 'Region')
display(vgSalesMelt)
```

```python
vgRegion = pd.read_csv('VideoGameRegion.csv', keep_default_na = False)
display(vgRegion)
```

```python
vgSalesMerge = vgSalesMelt.merge(vgRegion, left_on = 'Region', right_on = 'Abbreviation')
display(vgSalesMerge)
```

```python
vgSalesMerge.drop(columns = ['Name', 'Abbreviation', 'Region'], inplace = True)
```

```python
display(vgSalesMerge)
```

```python
vgSalesMerge = vgSalesMerge[['NameKey', 'PublisherKey', 'DeveloperKey', 'RegionKey', 'Sales']]
display(vgSalesMerge)
```

```python
vgSalesMerge.sort_values(by = ['NameKey', 'RegionKey'], inplace = True)
display(vgSalesMerge)
```

```python
vgSalesMerge.to_csv('FactSales.csv', index = False)
```
